```python
# import libraries
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt
from google.colab import files
import io
import datetime
```

```python
# load image
uploaded = files.upload()
```

```python
'''Change the file name as the image file u will be uploading in previous line
Read the img file'''
img = cv2.imread('happy.jpeg')
plt.imshow(img[:,:,::-1])
plt.show()
```

```python
# analyze
result = DeepFace.analyze(img,actions=['emotion'])
```

```python
# print the result
print(result)
```

```python
# print the emtion
print(result['dominant_emotion'])
```

```python
# save the emtion in a textfile
filename= datetime.datetime.now()
with open(filename.strftime("%d %B %Y")+".txt", "w") as file: 
  file.write(result['dominant_emotion'])			
file.close()
```
